```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
path = "/content/drive/MyDrive/AIML/online_shoppers_intention.csv"
data = pd.read_csv(path)
data.head()
```

```python
most_purchased_product = data['ProductRelated'].idxmax()
print("Most Liked and Purchased Product (Index):", most_purchased_product)
```

```python
special_days = data['SpecialDay'].value_counts()
print("Purchases by Special Days:\n", special_days)
```

```python
top_days = data.groupby(['SpecialDay', 'VisitorType']).size().reset_index(name='Counts')
top_5_days = top_days.nlargest(5, 'Counts')
print("Top 5 Special Days and Visitor Types:\n", top_5_days)
```

```python
monthly_purchases = data['Month'].value_counts()
print("Months with Most Purchases:\n", monthly_purchases)
```

```python
total_revenue = data.loc[data['Revenue'] == True, 'PageValues'].sum()
print("Total Revenue from Purchases:", total_revenue)
```

```python

data['PageValues'] = pd.to_numeric(data['PageValues'], errors='coerce')
purchases_data = data[data['Revenue'] == True]
max_revenue = purchases_data['PageValues'].max()
min_revenue = purchases_data['PageValues'].min()
print("Maximum Revenue Generated:", max_revenue)
print("Minimum Revenue Generated:", min_revenue)
```
